```python
from textwrap import dedent

from agno.agent import Agent
from agno.models.groq import Groq
from agno.tools.firecrawl import FirecrawlTools
from agno.tools.reasoning import ReasoningTools

competitor_analysis_agent = Agent(
    model=Groq(id="llama-3.3-70b-versatile"),
    tools=[
        FirecrawlTools(
            enable_search=True,
            enable_crawl=True,
            enable_mapping=True,
            formats=["markdown", "links", "html"],
            search_params={
                "limit": 2,
            },
            limit=5,
        ),
        ReasoningTools(
            add_instructions=True,
        ),
    ],
    tool_choice="auto",
    instructions=[
        "Conduct a competitive analysis for the given company in the payments industry.",
        "Use Firecrawl tools to search, crawl, and map websites.",
        "Identify products, pricing, and market positioning.",
        "Perform SWOT analysis for competitors and synthesize recommendations.",
        "Always use the 'think' tool for planning and 'analyze' for processing findings.",

    ],
    expected_output=dedent("""\
    # Competitive Analysis Report: {Target Company}

    ## Executive Summary

    ## Research Methodology

    ## Market Overview
    ### Industry Context
    ### Competitive Landscape

    ## Competitor Analysis
    ### Competitor 1: {Name}
    #### Company Overview
    #### Products & Services
    #### Digital Presence Analysis
    #### SWOT Analysis

    ### Competitor 2: {Name}

    ### Competitor 3: {Name}

    ## Comparative Analysis
    ### Feature Comparison Matrix
    ### Pricing Comparison
    ### Market Positioning Analysis

    ## Strategic Insights
    ### Key Findings
    ### Competitive Advantages
    ### Competitive Risks

    ## Strategic Recommendations
    ### Immediate Actions (0-3 months)
    ### Short-term Strategy (3-12 months)
    ### Long-term Strategy (12+ months)

    ## Conclusion
    """),
    markdown=True,
    add_datetime_to_context=True,
    stream_events=True,
)


competitor_analysis_agent.print_response(
    """Analyze the competitive landscape for Stripe in the payments industry.
   """,
    stream=True,
    show_full_reasoning=True,
)
'''
competitor_analysis_agent.print_response(
    """Analyze the competitive landscape for Stripe in the payments industry.
    Focus on their products, pricing models, and market positioning.""",
    stream=True,
    show_full_reasoning=True,
)'''
```
